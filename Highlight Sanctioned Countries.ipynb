```python
import folium
import geopandas as gpd

# Load the Natural Earth dataset manually
world = gpd.read_file("C:\\Users\\akaus\\OneDrive\\Desktop\\Anaconda\\Ad\\ne_110m_admin_0_countries.shp")  # Replace with the actual path where file is kept

# List of countries to color red
countries_to_highlight = [
    "Afghanistan", "Belarus", "Burma", "Cuba", "North Korea", "Iran", "Iraq", "Libya",
    "Russia", "South Sudan", "Sudan", "Syria", "Ukraine", "Venezuela", "Yemen"
]

# Filter the dataset for the specified countries
selected_countries = world[world["NAME"].isin(countries_to_highlight)]  # Ensure column name matches your dataset

# Create a map centered around a reasonable location
m = folium.Map(location=[20, 50], zoom_start=3, tiles="CartoDB Positron")

# Add selected countries to the map with red fill
folium.GeoJson(
    selected_countries,
    style_function=lambda x: {"fillColor": "red", "color": "black", "weight": 1.5, "fillOpacity": 1.0},
).add_to(m)

# Save the map to an HTML file
m.save("highlighted_countries_map.html")

# Display the map (for Jupyter Notebook users)
m
```
